```python
import pandas as pd
import numpy as np
```

```python
#Review data
df = pd.read_csv("Data/Accidental_Drug_Related_Deaths_2012-2018.csv")
df.head()
```

```python
# Review counts of all accident related deaths
df.count()
```

```python
df.columns
```

```python
# Review Lat/Long info
df[['DeathCity','ResidenceCity','InjuryCity','DeathCityGeo','ResidenceCityGeo', 'InjuryCityGeo']].head()
```

```python
# Clean up Geo columns - remove names and keep just lat/long
import re

df['DeathCityGeo'] =df.loc[:,'DeathCityGeo'].str.extract('.*\((.*)\).*')
df['ResidenceCityGeo'] =df.loc[:,'ResidenceCityGeo'].str.extract('.*\((.*)\).*')
df['InjuryCityGeo'] =df.loc[:,'InjuryCityGeo'].str.extract('.*\((.*)\).*')
df.head()
```

```python
# Split out DeathCityGeo coordinates into latitude and longitude columns
df['DeathCityLat'] = df.DeathCityGeo.str.split(', ').apply(lambda x: x[0])
df['DeathCityLong'] = df.DeathCityGeo.str.split(', ').apply(lambda x: x[1])


df.head()
```

```python
# Assign Fentanyl/FentanylAnalogue/OpiateNOS columns to Synthetic Opioid Category

cols = ['Fentanyl', 'FentanylAnalogue','OpiateNOS']
df["Synthetic Opioids"] = (df[cols]=="Y").any(axis="columns")
df["Synthetic Opioids"].value_counts()
```

```python
# Assign Oxycodone', 'Oxymorphone','Hydrocodone','Methadone','Morphine_NotHeroin','Hydromorphone' columns 
# to Prescription Opioid Category

cols2 = ['Oxycodone', 'Oxymorphone','Hydrocodone','Methadone','Morphine_NotHeroin','Hydromorphone']
df['Prescription Opioids'] = (df[cols2]=="Y").any(axis="columns")
df['Prescription Opioids'].value_counts()
```

```python
# Replace Y/NaN in Heroin column with True/False to be consistent with other category columns

df["Heroin"].fillna("N", inplace = True) 
df =df.replace({'Heroin': {'Y': True, 'N': False}})

df["Heroin"].value_counts()
```

```python
df['Race'].value_counts()
```

```python
# Clean up Race column and put into categories of White, Black, Asian, Hispanic, Other
df['Race'] = df['Race'].replace(
    {'Hispanic, White': 'Hispanic', 'Hispanic, Black': 'Hispanic','Asian, Other':'Asian','Asian Indian':'Asian','Chinese':'Asian',
    'Native American, Other':'Other','Hawaiian':'Other'})
df['Race'].value_counts()
```

```python
# Clean up city data
frequencies = df['DeathCity'].value_counts()

condition = frequencies<20   
mask_obs = frequencies[condition].index
mask_dict = dict.fromkeys(mask_obs, 'Other')

df['DeathCity'] = df['DeathCity'].replace(mask_dict) 
df['DeathCity'].value_counts()
```

```python
# Create a new dataframe with only columns needed
clean_df = df[['ID', 'Date', 'Age', 'Sex', 'Race', 'ResidenceCity',
       'ResidenceCounty', 'ResidenceState', 'DeathCity', 'DeathCounty',
       'Location', 'LocationifOther', 'DescriptionofInjury', 'InjuryPlace',
       'InjuryCity', 'InjuryCounty', 'InjuryState', 'COD',
       'Heroin','Synthetic Opioids','Prescription Opioids','DeathCityGeo',
       'DeathCityLat','DeathCityLong'
       ]]
clean_df.head()
```

```python
# Create a new column for Drug Category

def set_drug(row):
    if row["Heroin"] == True:
        return "Heroin"
    elif row["Synthetic Opioids"] == True:
        return "Synthetic Opioids"
    else:
        return "Prescription Opioids"

clean_df = clean_df.assign(Drug_Category=clean_df.apply(set_drug, axis=1))

clean_df.head()


```

```python
# Create a new dataframe with only deaths related to the 3 categories of Opioid deaths excluding Cocaine, Ethanol, and Benxodiazeprine
opioid_df = clean_df.loc[(clean_df['Heroin']== True)|(clean_df['Synthetic Opioids']==True)|(clean_df['Prescription Opioids'] == True), :]

opioid_df.count()
```

```python
# Push dataframe to csv to be used for machine learning and visualizations
opioid_df.to_csv("Opioid_Related_Deaths_2012-2018.csv",
                  encoding="utf-8", index=False, header=True)
```
