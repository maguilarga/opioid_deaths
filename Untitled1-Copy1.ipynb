```python
#Dependencies: 
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime as dt
import time 
import calendar
from numpy import array
```

```python
#Import Data: 
df=pd.read_csv('./Data/Accidental_Drug_Related_Deaths_2012-2018_Clean.csv')
df.head()
```

```python
#Fix the date format:
df["Date"] = pd.to_datetime(df["Date"])
df.head()
```

```python
#Make the total death column:
df['ID']=df["ID"].str.replace('-', '')
df['ID']=df['ID'].astype('int64', errors='ignore')
df.ID.head()
```

```python
df['Date'] = df.Date.map(dt.toordinal)
```

```python
ser=pd.Series(index='Date', data='ID')
ser_df=ser.to_frame()

ser_df.reset_index(inplace=True)
ser_df.columns=['Years', 'Deaths']
ser_df.plot(kind='scatter', x='Date', y='ID')
plt.show()
```
